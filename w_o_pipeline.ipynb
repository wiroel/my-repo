```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
from google.colab import drive
drive.mount('/content/drive')

file_path="/content/drive/MyDrive/titanic.csv"

df = pd.read_csv(file_path)

df
```

```python
print(df.columns)
print(df)
```

```python
df.drop(columns=['Name', 'Cabin', 'PassengerId', 'Ticket'], inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']), df['Survived'], test_size=0.2, random_state=42)
```

```python
df.isnull().sum()
```

applying simple imputer

```python
si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy = 'most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.fit_transform(X_test[['Age']])
X_test_embarked = si_embarked.fit_transform(X_test[['Embarked']])
```

```python
X_train_age
```

applying OHE

handle-UNknown = This parameter decides what to do if test data contains categories not seen during training.

sparse_output=This controls whether the output matrix is stored as a sparse matrix or a dense array.



```python
ohe_sex = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.fit_transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.fit_transform(X_test_embarked)
```

```python
X_train_embarked
```

```python
X_train.head(2)
```

```python
X_train_rem = X_train.drop(columns=['Sex', 'Embarked', 'Age'])

X_test_rem = X_test.drop(columns=['Embarked', 'Sex', 'Age'])
```

```python
X_train_transformed = np.concatenate((X_train_rem, X_train_age, X_train_sex, X_train_embarked), axis=1)

X_test_transfromed = np.concatenate((X_test_rem, X_test_age, X_test_sex, X_test_embarked), axis=1)
```

```python
X_train_transformed.shape
```

Decision Tree Classifier

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed, y_train)
```

```python
y_pred = clf.predict(X_test_transfromed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
